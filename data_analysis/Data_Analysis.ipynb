```python
# Lấy văn bản từ file txt với mã hóa khác
file_path = "E:/HealthSymptoChat/dataset/sosuckhoe.txt"
with open(file_path, "r", encoding="utf-8") as file:
    content = file.read()
```

```python
from collections import Counter
split_it = content.split()
Counter = Counter(split_it)
most_occur = Counter.most_common(20)

print(most_occur)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.DataFrame(most_occur, columns=['Từ vựng', 'Tần suất xuất hiện'])
sns.barplot(x='Tần suất xuất hiện', y='Từ vựng',
            data=df, orient="h",
            palette=("husl"))
plt.title("Top 20 từ có tần suất xuất hiện cao nhất")
plt.show()
```

```python
from collections import Counter
bigrams = zip(split_it, split_it[1:])
counts = Counter(bigrams)
most_occur_bigrams = counts.most_common(20)
print(most_occur_bigrams)
```

```python
df = pd.DataFrame(most_occur_bigrams, columns=['Cặp từ', 'Tần suất xuất hiện'])
for i in range(len(df)):
  df["Cặp từ"][i]=df["Cặp từ"][i][0]+" "+df["Cặp từ"][i][1]
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.barplot(x='Tần suất xuất hiện', y='Cặp từ',
            data=df, orient="h",
            palette=("husl"))
plt.title("Top 20 cặp từ có tần suất xuất hiện cao nhất")
plt.show()
```

```python
import collections

def trigram_frequency(text):
    # Split the text into words
    words = text
    # Create a dictionary to store trigram counts
    trigram_counts = collections.Counter()

    # Iterate over the words and create trigrams
    for i in range(2, len(words)):
        trigram = words[i - 2] + " " + words[i - 1] + " " + words[i]
        trigram_counts[trigram] += 1

    return trigram_counts

# Example usage
most_occur_trigram = trigram_frequency(split_it).most_common(20)

# Print the most frequent trigrams
print(most_occur_trigram)
```

```python
df = pd.DataFrame(most_occur_trigram, columns=['Bộ ba từ vựng', 'Tần suất xuất hiện'])
```

```python
sns.barplot(x='Tần suất xuất hiện', y='Bộ ba từ vựng',
            data=df, orient="h",
            palette=("husl"))
plt.title("Top 20 bộ ba từ vựng có tần suất xuất hiện cao nhất")
plt.show()
```
